### AI tools next year

#### All Respondents
https://survey.stackoverflow.co/2023/#developer-tools-ai-next

```python
ai_workflow = ['AINextVery different', 'AINextSomewhat different', 'AINextNeither different nor similar', 'AINextSomewhat similar', 'AINextVery similar']

# list of unique options from AINextVery different
ai_tools_nxt = []
for items in results_df['AINextSomewhat different'].dropna().unique():
    for item in items.split(';'):
        if item in ai_tools_nxt:
            pass
        else:
            ai_tools_nxt.append(item)

info_df_counts = pd.DataFrame(index=ai_tools_nxt, columns=ai_workflow)
```

```python
for tool in ai_tools_nxt:
    for item in ai_workflow:
        val = results_df.loc[results_df[item].str.contains(tool, na=False, regex=False), item].count()
        info_df_counts.loc[tool, item] = val

# info_df_counts
```

```python
info_df_percentage = info_df_counts.copy()

for tool in ai_tools_nxt:
    total = info_df_percentage.loc[tool].sum()
    for item in ai_workflow:
        val = info_df_percentage.loc[tool, item] / total * 100
        info_df_percentage.loc[tool, item] = f'{val:.2f} %'
        
info_df_percentage
```
